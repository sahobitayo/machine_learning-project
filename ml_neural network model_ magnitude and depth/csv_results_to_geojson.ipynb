```python
# Import Libraries

import pandas as pd
import numpy as np
import geojsonio
import geopandas as gp
import numpy as np
import csv
import json
from geojson import Feature, FeatureCollection, Point
```

```python
# Read in raw data from csv
data = pd.read_csv('ml_results.csv')
data
```

```python
# columns used for constructing geojson object
features = data.apply(
    lambda row: Feature(geometry=Point((float(row['longitudes']), float(row['latitudes'])))),
    axis=1).tolist()

# all the other columns used as properties
properties = data.drop(['longitudes', 'latitudes'], axis=1).to_dict('records')

# whole geojson object
feature_collection = FeatureCollection(features=features, properties=properties)
```

```python
with open('ml_neuron_model_results.geojson', 'w', encoding='utf-8') as f:
    json.dump(feature_collection, f, ensure_ascii=False)
```
